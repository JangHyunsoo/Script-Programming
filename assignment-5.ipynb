    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

    > \>\>\> c = Counter(10) <br/>
    > \>\>\> d = Counter(10, 2) <br/>
    
  - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 \_\_str\_\_() 메소드를 Counter 클래스 내에 구현하시오.

    > \>\>\> c.incr()  <br/>

  - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오. 
    > \>\>\> c()  <br/>
    > \>\>\> d()  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 12 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 14 <br/>
    
  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 17 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 9 <br/>  
    
  - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> print(c > 10)  <br/>
    > True <br/> 
    > \>\>\> print(d > 10) <br/>
    > False <br/>
    > \>\>\> print(c < 10)  <br/>
    > False <br/> 
    > \>\>\> print(d < 10) <br/>
    > True <br/>
    > \>\>\> print(c == 17)  <br/>
    > True <br/> 
    > \>\>\> print(d != 9) <br/>
    > False <br/>

````python
class Counter:
    def __init__(self, count = 1, step = 1):
        self.count = count
        self.step = step
        
    def incr(self):
        self.count += self.step
        
    def __str__(self):
        return "[Count (step: {0})] {1}".format(self.step, self.count)
    
    def __call__(self):
        self.incr()
        
    def __add__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count + value, self.step)
        return Counter(self.count, self.step)
    
    def __sub__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count - value, self.step)
        return Counter(self.count ,self.step)
    
    def __radd__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count + value, self.step)
        return Counter(self.count, self.step)
    
    def __rsub__(self, value):
        if(isinstance(value, int)):
            return Counter(value - self.count, self.step)
        return Counter(self.count ,self.step)
    
    def __lt__(self, other):
        if(isinstance(other, int)):
            return self.count > other
        elif(isinstance(other, Counter)):
            return self.count > other.count 
        return False
    
    def __le__(self, other):
        if(isinstance(other, int)):
            return self.count >= other
        elif(isinstance(other, Counter)):
            return self.count >= other.count 
        return False
    
    def __eq__(self, other):
        if(isinstance(other, int)):
            return self.count == other
        elif(isinstance(other, Counter)):
            return self.count == other.count 
        return False
    
c = Counter(10)
d = Counter(10, 2)

print(c != 10)
print(d)
````

### 설명
1. `__init__`을 오버로딩하여 생성자를 정의하였고 count와 step의 default는 모두 1로 하였다.
2. `__str__`를 오버로딩하였고 format을 이용하여 Counter를 str로 바꿔주는 내용이다.
3. incr는 새로 정의하였고 self.count += self.step을 이용하여 count를 증가시켜주었다.
4. `__call__`을 오버로딩하여 위의 c()처럼 사용할 수 있게 만들었고 내부적으로는 self.incr()를 불러 증가시켜주었다.
5. (+,-)를 오버로딩하기 위하여 `__sub__`와 `__add__`를 오버로딩하였고 추가적으로 `__rsub__`와 `__radd__`도 정의해주었다.
6. (==, >, <, >=, <=)를 오버로딩하기 위하여 `__it__`, `__le__`, `__eq__`을 오버도딩해주었다. 이유는 `__it__`을 만들면 `__gt__`가 `__le__`를 만들면 `__ge__`가 `__eq__`를 정의하면 `__nq__`가 추가적으로 얻을 수 있기 때문에 모두 정의하지 않았다.

## Problem 4

4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

````python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
````

### 설명
0. 위 클래스는 기존에 있는 list를 상속받아 MySet를 만드는 함수이다. 때문에 list에 있는 기본적인 메소드들을 사용할 수 있다.
1. `__init__`은 생성자로 클래스를 통하여 객체를 만들때 불러지는 매직 메소드이다. 현재 클래스의 경우 리스트를 받아 이를 사진의 리스트에 넣고 후에 나온는 eliminate_duplicate을 이용하여 중복된 요소를 지우는 역활을 한다.
2. `__str__`은 str()을 이용하여 객체를 부를 경우 문자열로 바꾸기 위하여 불러지는 매직 메소드이다. 여기서는 처음에는 result에 "MySet: {"을 넣어주고 여기에 요소를 하나씩 넣어준다. 이후 마지멕에 들어간 " ,"을 지워주고 "}"를 넣어주기 위하여 result[0:len(result)-2] + "}"를 넣우다. 마지막으로 위에서 만든 result 문자열을 반환해준다.
3. 마지막은 eliminate_duplicate이다 이는 위와 다르게 일반적인 정적 메소드이며 정적 메소드는 현재 객체가 없어도 사용할 수 있는 객체이다. 때문에 MySet.eliminate_duplicate(l) 처럼 사용할 수 도 있다. 함수를 설명하자면 self를 받지 않기 때문에 인자로 list를 받아 사용하였다. 함수 안으로 들어가면 임시 객체인 s에 l의 요소를 넣는다. 이때 현재까지의 s에 들어않은 요소만 넣어 중복을 재거한다. 마지막으로는 l에 s를 복사해주고 함수를 마친다.

## Problem 5

5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

  - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> u = s | t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> u = s & t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {2, 3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> u = s - t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2} <br/>

````python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    
    def __or__(self, other):
        s = list()
        for e in self:
            s.append(e)
        for e in other:
            s.append(e)
        return MySet(s)
    
    def __and__(self, other):
        s = list()
        for e in other:
            if e in self:
                s.append(e)
        return MySet(s)
    
    def __sub__(self, other):
        s = list()
        for e in self:
            if e not in other:
                s.append(e)
        return MySet(s)
    
if __name__ == "__main__":
    s = MySet([1, 2, 3])
    t = t = MySet([3, 4, 5])
    u = s - t
    print(u)
````

### 설명
0. 중복은 생성을 하며 제거가 됨으로 따로 해주지 않았습니다.
1. `|`는 `__and__`를 재정의하여 사용하였고 함수 내용은 other의 요소 중 self에 요소가 중복되어 있다면 그것을 s에 넣어두고 최종적으로 나온 s를 이용하여 MySet을 생성하였다.
2. `&`는 `__or__`를 재정의하여 사용하였고 함수 내용은 other과 self의 요소를 모두 연결하여 s에 넣어두고 최종적으로 나온 s를 이용하여 MySet을 생성하였다.
3. `-`는 `__sub__`를 재정의하여 사용하였고 함수 내용은 self의 요소 중 other에 있는 요소가 중복되어 있지 않다면 그것을 s에 넣어두고 최종적으로 나온 s를 이용하여 MySet을 생성하였다.

````python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def getWordsFrequency(self, reverse=False):
        total_frequencys = super().getWordsFrequency()
        total_frequencys = {k:v for k,v in sorted(list(total_frequencys.items()), key=lambda x: x[1], reverse= not reverse)}
        return total_frequencys
    
    def __iter__(self):
        self.iter_total_frequencys = [item for item in self.getWordsFrequency().items()]
        self.iter_curr = 0
        self.iter_end = len(self.iter_total_frequencys)
        return self
    
    def __next__(self):
        if(self.iter_curr < self.iter_end):
            result = self.iter_total_frequencys[self.iter_curr]
            self.iter_curr += 1
            return result
        else:
            raise StopIteration

w1 = SearchEngineWithOrderedWebWords("https://animal-crossing.com/", "https://www.python.org/", "https://www.notion.so/", "https://aws.amazon.com/", "https://www.billboard.com/")

# print(w1.getWordsFrequency())
 
# for i in w1:
#     print(i)
````

### 설명
`SearchEngine`
1. 생성자는 `__init__`를 통하여 만들었고 url를 `*args`를 통하여 받아 아래에서 설명한 addUrl를 통하여 하나씩 추가해주었다.
2. `addUrl`은 urls에 str형태로 url이 없으면 추가해주고 이에 맞는 freq를 dictionary<url, freq>의 형태로 frequencys에 넣어주었다. 이때 과거에 frequency를 만들어주는 함수를 한 번에 부르기 위하여 makeFrequency를 static함수로 만들고 호출해주었다.
3. `removeUrl`도 마찬가지로 urls에서 url이 있으면 지워주고 frequencys에서도 삭제해준다.
4. 개인적으로 만든 updateSource는 freq 즉 웹의 데이터가 바뀐 것을 적용하기 위하여 update해주는 함수이다.
5. `listUrl`은 urls의 데이터를 출력해준다.
6. `getWordsFrequency`는 위에서 생성된 frequencys를 전부 더하여 만들어주었다.
7. `getMaxWordsFrequency`는 위의 `getWordsFrequency` 부른 뒤 items로 바꾼 후 sorting하여 0번 인덱스의 값과 같은 값을 dictonary에 넣어 반환해주었다.
8. `searchUrlByWord`는 이전에 만든 search에 관련된 함수를 통하여 유사도를 검사하여 정보들 만들어내고 이 (유사도, 빈도, 문자) 중 빈도가 가장 큰것과 같은 url를 출력하기 위하여 sorting 후 0번 인덱스와 비교하였다. 이때 staticmethod를 만들어 이전에 문자 자르기, 번역하기, 불사전 처리 등을 한 후 유사도를 계산하여 반환해준다.

`SearchEngineWithOrderedWebWords`
1. `__init__`은 바뀌는게 없어 따로 overriding하지 않았다.
2. `getWordsFrequency`는 인자로 reserve를 주었고 기본으로는 False이다. 함수 내용은 super에 `getWordsFrequency`를 불러 total_freq를 생성해주고 이를 reverse에 따라 sorting하여 반환해주었다.
3. 마지막으로 iterator를 사용할 수 있게 바꾸기 위하여 `__iter__`에서는 iter_total_frequencys에 getWordsFrequency()를 items를 통하여 만들어 주었고 현제 보고 있는 값을 알기위하여 iter_curr라는 idx를 생성해주었다. iter_end에는 iter_total_frequencys의 크기를 넣어 언제 오류를 발산하면 되는지 알려주었다. `__next__`에는 현제 iter_curr의 값을 리턴해주고 iter_curr를 하나씩 증가시킨다. 만약 iter_curr가 iter_end와 같아지면 오류를 발산해주었ㄷ.
