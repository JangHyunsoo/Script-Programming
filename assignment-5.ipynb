    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

    > \>\>\> c = Counter(10) <br/>
    > \>\>\> d = Counter(10, 2) <br/>
    
  - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 \_\_str\_\_() 메소드를 Counter 클래스 내에 구현하시오.

  - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오. 
    > \>\>\> c()  <br/>
    > \>\>\> d()  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 12 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 14 <br/>
    
  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 17 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 9 <br/>  
    
  - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> print(c > 10)  <br/>
    > True <br/> 
    > \>\>\> print(d > 10) <br/>
    > False <br/>
    > \>\>\> print(c < 10)  <br/>
    > False <br/> 
    > \>\>\> print(d < 10) <br/>
    > True <br/>
    > \>\>\> print(c == 17)  <br/>
    > True <br/> 
    > \>\>\> print(d != 9) <br/>
    > False <br/>

```python
class Counter:
    def __init__(self, count = 1, step = 1):
        self.count = count
        self.step = step
        
    def incr(self):
        self.count += self.step
        
    def __str__(self):
        return "[Count (step: {0})] {1}".format(self.step, self.count)
    
    def __call__(self):
        self.incr()
        
    def __add__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count + value, self.step)
        return Counter(self.count, self.step)
    
    def __sub__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count - value, self.step)
        return Counter(self.count ,self.step)
    
    def __radd__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count + value, self.step)
        return Counter(self.count, self.step)
    
    def __rsub__(self, value):
        if(isinstance(value, int)):
            return Counter(self.count - value, self.step)
        return Counter(self.count ,self.step)
    
    def __lt__(self, other):
        if(isinstance(other, int)):
            return self.count > other
        elif(isinstance(other, Counter)):
            return self.count > other.count 
        return False
    
    def __le__(self, other):
        if(isinstance(other, int)):
            return self.count >= other
        elif(isinstance(other, Counter)):
            return self.count >= other.count 
        return False
    
    def __eq__(self, other):
        if(isinstance(other, int)):
            return self.count == other
        elif(isinstance(other, Counter)):
            return self.count == other.count 
        return False
    
c = Counter(10)
d = Counter(10, 2)

print(c)
print(d)
```

### 설명
1. __init__을 오버로딩하여 생성자를 정의하였고 count와 step의 default는 모두 1로 하였다.
2. __str__를 오버로딩하였고 format을 이용하여 Counter를 str로 바꿔주는 내용이다.
3. incr는 새로 정의하였고 self.count += self.step을 이용하여 count를 증가시켜주었다.
4. __call__을 오버로딩하여 위의 c()처럼 사용할 수 있게 만들었고 내부적으로는 self.incr()를 불러 증가시켜주었다.
5. (+,-)를 오버로딩하기 위하여 __sub__와 __add__를 오버로딩하였고 추가적으로 __rsub__와 __radd__도 정의해주었다.
6. (==, >, <, >=, <=)를 오버로딩하기 위하여 __it__, __le__, __eq__을 오버도딩해주었다. 이유는 __it__을 만들면 __gt__가 __le__를 만들면 __ge__가 __eq__를 정의하면 __nq__가 추가적으로 얻을 수 있기 때문에 모두 정의하지 않았다.

## Problem 4

4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

### 설명
0. 위 클래스는 기존에 있는 list를 상속받아 MySet를 만드는 함수이다. 때문에 list에 있는 기본적인 메소드들을 사용할 수 있다.
1. __init__은 생성자로 클래스를 통하여 객체를 만들때 불러지는 매직 메소드이다. 현재 클래스의 경우 리스트를 받아 이를 사진의 리스트에 넣고 후에 나온는 eliminate_duplicate을 이용하여 중복된 요소를 지우는 역활을 한다.
2. __str__은 str()을 이용하여 객체를 부를 경우 문자열로 바꾸기 위하여 불러지는 매직 메소드이다. 여기서는 처음에는 result에 "MySet: {"을 넣어주고 여기에 요소를 하나씩 넣어준다. 이후 마지멕에 들어간 " ,"을 지워주고 "}"를 넣어주기 위하여 result[0:len(result)-2] + "}"를 넣우다. 마지막으로 위에서 만든 result 문자열을 반환해준다.
3. 마지막은 eliminate_duplicate이다 이는 위와 다르게 일반적인 정적 메소드이며 정적 메소드는 현재 객체가 없어도 사용할 수 있는 객체이다. 때문에 MySet.eliminate_duplicate(l) 처럼 사용할 수 도 있다. 함수를 설명하자면 self를 받지 않기 때문에 인자로 list를 받아 사용하였다. 함수 안으로 들어가면 임시 객체인 s에 l의 요소를 넣는다. 이때 현재까지의 s에 들어않은 요소만 넣어 중복을 재거한다. 마지막으로는 l에 s를 복사해주고 함수를 마친다.

## Problem 5

5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

  - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> u = s | t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> u = s & t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {2, 3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> u = s - t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    
    
    def __and__(self, other):
        s = list()
        for e in self:
            s.append(e)
        for e in other:
            s.append(e)
        return MySet(s)
    
    def __or__(self, other):
        s = list()
        for e in other:
            if e in self:
                s.append(e)
        return MySet(s)
    
    def __sub__(self, other):
        s = list()
        for e in self:
            if e not in other:
                s.append(e)
        return MySet(s)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s & t
    print(u)
```

### 설명
0. 중복은 생성을 하며 제거가 됨으로 따로 해주지 않았습니다.
1. `|`는 __or__를 재정의하여 사용하였고 함수 내용은 other의 요소 중 self에 요소가 중복되어 있다면 그것을 s에 넣어두고 최종적으로 나온 s를 이용하여 MySet을 생성하였다.
2. `&`는 __and__를 재정의하여 사용하였고 함수 내용은 other과 self의 요소를 모두 연결하여 s에 넣어두고 최종적으로 나온 s를 이용하여 MySet을 생성하였다.
3. `-`는 __sub__를 재정의하여 사용하였고 함수 내용은 self의 요소 중 other에 있는 요소가 중복되어 있지 않다면 그것을 s에 넣어두고 최종적으로 나온 s를 이용하여 MySet을 생성하였다.

```python
import requests

def sub(s, l, r):
    origin_l, tmp1 = s.split(l,1)
    tmp2, origin_r = tmp1.split(r,1)
    return origin_l + " " + origin_r
    

def removeHtmlTag(source):
    
    # 주석 삭제
    while source.find("//<") != -1:
        source = sub(source, "//<", ">")

    # 요소내부에서 앵커요소 나오는 경우가 있는데 어떻게 처리해야할지 모르겠습니다.
    # 예시. <a href=\" < url > https://gallery.v.daum.net/p/premium/excavation/cSCjAyizEy</url>\" class=\"link_thumb photo_pic1\" data-tiara-action-name=\"오늘의 포토_데이터 운영_하단 썸네일\" data-tiara-layer=\"thumbnail\">
    while source.find("<url>") != -1:
        source = sub(source, "<url>","</url>")
        
    while source.find("<text>") != -1:
        source = sub(source, "<text>","</text>")
        
    # <script> 삭제
    while source.find("<script") != -1:
        source = sub(source, "<script", "</script>")
    
    # <style> 삭제
    while source.find("<style") != -1:
        source = sub(source, "<style", "</style>")

    cursor = source

    while cursor.find("<") != -1:
        cursor = sub(cursor,"<",">")
    
    cursor = cursor.replace("\t","")
    cursor = cursor.replace("\n","")
    cursor = cursor.replace("\r","")
    result = cursor.split(' ')
    
    for _ in range(result.count('')):
        result.remove('')
    
    return result
```

```python
import functools

# 가중치 생성
def makeDistWithString(key_origin, s):
    dict_ilst = [1 / float(2 ** (i + 1)) for i in range(len(key_origin))]
    max_dict = 0
    lower_s = lower_all(s)
    
    for i in range(len(key_origin)):
        key = key_origin[i:]
        now_dict = functools.reduce(lambda x,y: x+y,dict_ilst[i:])
        idx = lower_s.find(lower_all(key))
        cal_len = float(len(key)) / len(s)
        cal_idx = 1 / (1 + idx)  if idx != -1 else 0
        cal_result = float(cal_idx) * float(cal_len) * float(now_dict)
        max_dict = max_dict if max_dict > cal_result else cal_result
    
    for i in range(1, len(key_origin)):
        key = key_origin[:i]
        now_dict = functools.reduce(lambda x,y: x+y,dict_ilst[i:])
        idx = lower_s.find(lower_all(key))
        cal_len = float(len(key)) / len(s)
        cal_idx = 1 / (1 + idx)  if idx != -1 else 0 
        cal_result = float(cal_idx) * float(cal_len) * float(now_dict)
        max_dict = max_dict if max_dict > cal_result else cal_result
    
    return (s, max_dict)
    

def makeDist(key_origin, d):
    l = [k for k in d.keys()]
    max_dict = -1
    max_word = " "

    for s in l:
        now_word, now_dict = makeDistWithString(key_origin, s)
        if max_dict < now_dict:
            max_dict = now_dict
            max_word = now_word 
        
    return (max_dict, d[max_word], max_word)
```
